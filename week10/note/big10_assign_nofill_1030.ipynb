```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset('titanic')
df = pd.DataFrame(titanic)
print(df.head())
```

1.class 별로 승객 수를 나타내는 countplot을 구하시오(image 참조).
a-1.png

```python
sns.set_theme(style='whitegrid')
sns.countplot(data=titanic, x=titanic["class"])
```

2. class 별로 승객 수를 나타내는 countplot을 구하시오. (생존 여부('alive)'를 세부 분류로 함.)

```python
sns.set_theme(style='whitegrid')
sns.countplot(data=titanic, x=titanic["class"], hue='alive', palette=['red', 'blue'])
```

3. class 별 count를 이용하여 pie chart를 작성 하시오.(image 참조)
a-3.png

```python
import matplotlib.pyplot as plt
import seaborn as sns
result=df.groupby('class')['class'].count()
```
